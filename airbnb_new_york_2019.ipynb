```python
# import libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
pd.set_option('display.max_columns', 200) 
```

```python
df = pd.read_csv('AB_NYC_2019.csv', sep=',').copy()
df.info()
```

```python
df.head().T
```

```python
# drop unnecessary columns

df.drop(columns='host_name', inplace=True)
```

```python
# rename columns

df.rename(columns={
    'id': 'listing_id',
    'name': 'listing_name',
    'number_of_reviews': 'total_reviews',
    'last_review': 'last_review_date',
    'calculated_host_listings_count': 'host_total_listings',
    'availability_365': 'available_days_in_year',
    }, inplace=True)

df.tail()
```

```python
df = df[['listing_id', 'listing_name', 
         'available_days_in_year',
        'neighbourhood', 'neighbourhood_group', 
        'room_type', 'price','minimum_nights', 
        'total_reviews', 'last_review_date', 'reviews_per_month',
        'latitude', 'longitude',
        'host_id', 'host_total_listings',        
        ]]

df.head()
```

* review columns, values and data types

```python
df.describe(include='all').T
```

#### first insights:

* listing_id values are unique for each rows but unique listing_name are less
* average available days of a property is 45 days in a year
* minimum_nights has some extremely high values up to 1250
* host_total_listings max value is 327

```python
df.duplicated().sum()
```

```python
df.isna().sum()
```

```python
# view nan values in listing_name column

df_filtered = df[df['listing_name'].isna()]
df_filtered.head(25).T
```

* they represent real rooms only with missing names, these may not have been booked during the year
* for these rows, I will assign a default name such as 'Untitled Listing'

```python
# replace missing listing names with 'Untitled Listing'

df['listing_name'].fillna('Untitled Listing', inplace=True)
```

```python
# view 0 values in available_days_in_year column

df_filtered = df[df['available_days_in_year'] == 0]
df_filtered.head(25).T
```

* even they may represent valid records that were simply unavailable during the last year, they should be removed to avoid calculation issues and skewed results 

```python
df = df[df['available_days_in_year'] != 0].reset_index(drop=True)
```

```python
# view nan values in last_review_date column

df_filtered = df[df['last_review_date'].isna()]
df_filtered.head(25).T
```

```python
# check if the missing values in both columns are in the same rows

if df['last_review_date'].isna().equals(df['reviews_per_month'].isna()):
    print("yes they are same.")
else:
    print("nope, check it")
```

* these records should be kept because they may represent new or low-traffic rooms that were available but simply did not receive any reviews or even bookings

```python
# convert 'neighbourhood' and 'room_type' to category

df['neighbourhood'] = df['neighbourhood'].astype('category')
df['room_type'] = df['room_type'].astype('category')
```

```python
df.isna().sum()
```

```python
df.describe(include='all').T
```

```python
# review price column with 0 values 

df_filtered = df[df['price'] == 0]
df_filtered.info()
```

```python
df_filtered.head(10).T
```

```python
df = df[df['price'] != 0].reset_index(drop=True)
df.info()
```

```python
# review total_reviews column with 0 values 

df_filtered = df[df['total_reviews'] == 0]
df_filtered.info()
```

```python
df_filtered.tail(20).T
```

* rows seem to represent true listings without reviews
* ince we cannot evaluate limits for extremely high values in price, minimum_nights and host_total_listings from current data, these records will remain as they are
